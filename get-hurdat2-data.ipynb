```python
import requests
import csv
from io import StringIO

# HURDAT2 contains information from the atlantic basin and the north-east pacific in separate files.

# URL of the webpage containing the data
url = "https://www.aoml.noaa.gov/hrd/hurdat/hurdat2.html"         # Atlantic Basin     (hurdat2-A)

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Read the content of the response
    data = response.text
    
    # Use StringIO to mimic a file object
    csv_data = StringIO(data)
    
    # Specify the delimiter used in your CSV
    delimiter = ','  # Assuming it's a comma-separated file
    
    # Read the CSV into a list of lists
    csv_reader = csv.reader(csv_data, delimiter=delimiter)
    
    # Specify the path where you want to save the CSV file
    output_file = "hurdat2-A.csv"
    
    # Open the output file in write mode
    with open(output_file, 'w', newline='') as file:
        # Create a CSV writer object
        csv_writer = csv.writer(file)
        
        # Write each row from the webpage to the output CSV file
        for row in csv_reader:
            # Skip empty rows and rows containing "<"
            if row and not any("<" in cell for cell in row):
                csv_writer.writerow(row)
    
    print("CSV file saved successfully!")
else:
    print("Failed to fetch data from the URL.")

    
# URL of the webpage containing the data
url = "https://www.aoml.noaa.gov/hrd/hurdat/hurdat2-nepac.html"   # North-East Pacific (hurdat2-P)

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Read the content of the response
    data = response.text
    
    # Use StringIO to mimic a file object
    csv_data = StringIO(data)
    
    # Specify the delimiter used in your CSV
    delimiter = ','  # Assuming it's a comma-separated file
    
    # Read the CSV into a list of lists
    csv_reader = csv.reader(csv_data, delimiter=delimiter)
    
    # Specify the path where you want to save the CSV file
    output_file = "hurdat2-P.csv"
    
    # Open the output file in write mode
    with open(output_file, 'w', newline='') as file:
        # Create a CSV writer object
        csv_writer = csv.writer(file)
        
        # Write each row from the webpage to the output CSV file
        for row in csv_reader:
            # Skip empty rows and rows containing "<"
            if row and not any("<" in cell for cell in row):
                csv_writer.writerow(row)
    
    print("CSV file saved successfully!")
else:
    print("Failed to fetch data from the URL.")
```
